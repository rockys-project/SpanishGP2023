The following information is available per lap (one DataFrame column for each):

* Time (pandas.Timedelta): Session time when the lap time was set (end of lap)

* Driver (string): Three letter driver identifier

* DriverNumber (str): Driver number

* LapTime (pandas.Timedelta): Recorded lap time. Officially deleted lap times will not be deleted here. Deleting laps is currently not supported.

* LapNumber (float): Recorded lap number

* Stint (float): Stint number

* PitOutTime (pandas.Timedelta): Session time when car exited the pit

* PitInTime (pandas.Timedelta): Session time when car entered the pit

* Sector1Time (pandas.Timedelta): Sector 1 recorded time

* Sector2Time (pandas.Timedelta): Sector 2 recorded time

* Sector3Time (pandas.Timedelta): Sector 3 recorded time

* Sector1SessionTime (pandas.Timedelta): Session time when the Sector 1 time was set

* Sector2SessionTime (pandas.Timedelta): Session time when the Sector 2 time was set

* Sector3SessionTime (pandas.Timedelta): Session time when the Sector 3 time was set

* SpeedI1 (float): Speedtrap sector 1 [km/h]

* SpeedI2 (float): Speedtrap sector 2 [km/h]

* SpeedFL (float): Speedtrap at finish line [km/h]

* SpeedST (float): Speedtrap on longest straight (Not sure) [km/h]

* IsPersonalBest (bool): Flag that indicates whether this lap is the official personal best lap of a driver. If any lap of a driver is quicker than their respective personal best lap, this means that the quicker lap is invalid and not counted. For example, this can happen if the track limits were exceeded.

* Compound (str): Tyres event specific compound name: SOFT, MEDIUM, HARD, INTERMEDIATE, WET (the actual underlying compounds C1 to C5 are not differentiated).

* TyreLife (float): Laps driven on this tire (includes laps in other sessions for used sets of tires)

* FreshTyre (bool): Tyre had TyreLife=0 at stint start, i.e. was a new tire

* Team (str): Team name

* LapStartTime (pandas.Timedelta): Session time at the start of the lap

* LapStartDate (pandas.Timestamp): Timestamp at the start of the lap

* TrackStatus (str): A string that contains track status numbers for all track status that occurred during this lap. The meaning of the track status numbers is explained in fastf1.api.track_status_data(). For filtering laps by track status, you may want to use Laps.pick_track_status().

* Position (float): Position of the driver at the end of each lap. This value is NaN for FP1, FP2, FP3, Sprint Shootout, and Qualifying as well as for crash laps.

* Deleted (Optional[bool]): Indicates that a lap was deleted by the stewards, for example because of a track limits violation. This data is only available when race control messages are loaded.

* DeletedReason (str): Gives the reason for a lap time deletion. This data is only available when race control messages are loaded.

* FastF1Generated (bool): Indicates that this lap was added by FastF1. Such a lap will generally have very limited information available and information is partly interpolated or based on reasonable assumptions. Cases were this is used are, for example, when a partial last lap is added for drivers that retired on track.

* IsAccurate (bool): Indicates that the lap start and end time are synced correctly with other laps. Do not confuse this with the accuracy of the lap time or sector times. They are always considered to be accurate if they exist! If this value is True, the lap has passed as basic accuracy check for timing data. This does not guarantee accuracy but laps marked as inaccurate need to be handled with caution. They might contain errors which can not be spotted easily. Laps need to satisfy the following criteria to be marked as accurate:

      * not an inlap or outlap

      * set under green or yellow flag (the api sometimes has issues with data from SC/VSC laps)

      * is not the first lap after a safety car period (issues with SC/VSC might still appear on the first lap after it has ended)

      * has a value for lap time and all sector times

      * the sum of the sector times matches the lap time (If this were to ever occur, it would also be logged separately as a data integrity error. You usually don’t need to worry about this.)

```python
def lap_time_analysis(driver_name):
  driver_laps = race.laps.pick_driver(driver_name).reset_index()
  df = driver_laps[['Driver','LapNumber','LapTime','Stint','Compound','PitInTime','PitOutTime']]
  import plotly.express as px
  import pandas as pd
  from sklearn.linear_model import LinearRegression

  # Assuming you have a DataFrame named df with columns 'Driver', 'LapNumber', 'LapTime', 'Stint', and 'Compound'

  # Sort the DataFrame by LapNumber to ensure the correct order on the x-axis
  df = df.sort_values('LapNumber')

  # Convert timedelta64 lap time values to seconds
  df['LapTime'] = df['LapTime'].dt.total_seconds()

  # Define the colors for each compound
  color_mapping = {
      'SOFT': 'red',
      'MEDIUM': 'yellow',
      'HARD': 'white'
  }

  # Create a scatter plot using Plotly Express
  fig = px.scatter(df, x='LapNumber', y='LapTime', color='Compound',
                  color_discrete_map=color_mapping)

  # Set the x-axis and y-axis labels
  fig.update_layout(xaxis_title='Lap Number', yaxis_title='Lap Time (seconds)')

  fig.update_layout(
      plot_bgcolor='darkblue'  # Specify your desired background color here
  )

  # Show the plot
  fig.show()


  #Removing the first lap of the new stint and also the previous lap because the car will slow down inside the pit lane and it will act as an outlier while analysing the stint
  #The data of Lap 1 is also removed because the race starts in a standing start and the lap time will be slow because of it which will act as an outlier
  #The flying laps will give proper readings for the stint analysis
  df2 = df[(df['PitInTime'].isnull()) & (df['PitOutTime'].isnull())].copy()

  filtered_df = df2.sort_values('LapNumber')

  # Convert timedelta64 lap time values to seconds
  #df['LapTime'] = df['LapTime'].dt.total_seconds()

  # Define the colors for each compound
  color_mapping = {
      'SOFT': 'red',
      'MEDIUM': 'yellow',
      'HARD': 'white'
  }

  # Create a scatter plot using Plotly Express
  fig = px.scatter(filtered_df, x='LapNumber', y='LapTime', color='Compound',
                  color_discrete_map=color_mapping, title='Plot After Removing the PitInLaps & PitOutLaps')

  # Set the x-axis and y-axis labels
  fig.update_layout(xaxis_title='Lap Number', yaxis_title='Lap Time (seconds)')

  fig.update_layout(
      plot_bgcolor='darkblue'  # Specify your desired background color here
  )

  # Show the plot
  fig.show()


  # Assuming you have a DataFrame named df with columns 'Driver', 'LapNumber', 'LapTime', 'Stint', and 'Compound'

  # Get the unique stint numbers in the DataFrame
  unique_stints = filtered_df['Stint'].unique()

  # Create a dictionary to store the separate DataFrames for each stint
  stint_dataframes = {}

  # Iterate over the unique stint numbers
  for stint_number in unique_stints:
      # Filter the DataFrame for the current stint number
      stint_df = filtered_df[filtered_df['Stint'] == stint_number].copy()

      # Add the DataFrame to the dictionary with the stint number as the key
      stint_dataframes[stint_number] = stint_df.reset_index(drop=True)


  regression_data = pd.DataFrame(columns=['Stint', 'LapNumber', 'LapTime', 'Predicted','Compound'])

  # Iterate over the stint_dataframes dictionary
  for stint_number, stint_df in stint_dataframes.items():
      # Extract the features (LapNumber) and target variable (LapTime)
      X = stint_df['LapNumber'].values.reshape(-1, 1)
      y = stint_df['LapTime'].values
      compound = stint_df['Compound'].values
      # Fit a linear regression model
      model = LinearRegression()
      model.fit(X, y)

      # Predict the target variable using the fitted model
      y_pred = model.predict(X)

      # Create a DataFrame with the stint number, lap numbers, actual lap times, and predicted lap times
      stint_regression_data = pd.DataFrame({'Stint': [stint_number] * len(X),
                                            'LapNumber': X.flatten(),
                                            'LapTime': y,
                                            'Predicted': y_pred,
                                            'Compound': compound})

      # Append the stint regression data to the overall regression_data DataFrame
      regression_data = pd.concat([regression_data, stint_regression_data], ignore_index=True)

  # Plot the scatter plot with regression lines for all stints
  fig = px.scatter(regression_data, x='LapNumber', y='LapTime', color='Compound',color_discrete_map=color_mapping,
                  title='Scatter Plot with Regression Lines')

  # Add regression lines
  for stint_number, stint_df in stint_dataframes.items():
      regression_line = regression_data[regression_data['Stint'] == stint_number]
      fig.add_trace(px.line(regression_line, x='LapNumber', y='Predicted', color='Compound',color_discrete_map=color_mapping).data[0])

  fig.update_layout(
      plot_bgcolor='darkblue', yaxis = dict(showgrid=False)  # Specify your desired background color here
  )
  fig.show()


  stint = 1
  print ('Average lap Times in Each Stint')
  print ('**Note We have excluded the PitInLap and the PitOutLap in the stint length')
  print ("")
  print ("")
  for i in stint_dataframes:
    print('STINT', stint, stint_dataframes[i]['LapTime'].mean(), 'Compound' , stint_dataframes[i]['Compound'][0], '   Length:', stint_dataframes[i]['LapTime'].count(), 'laps')
    stint = stint +1
```

```python
def compound_data(compound):
  race_df = race.laps.reset_index()

  race_df['LapTime'] = race_df['LapTime'].dt.total_seconds()

  race2_df = race_df[(race_df['PitInTime'].isnull()) & (race_df['PitOutTime'].isnull())].copy()


  drivers = race_df['Driver'].unique()

  driver_df = {}

  # Iterate over the unique stint numbers
  for driver in drivers:
      # Filter the DataFrame for the current stint number
      driver_df_temp = race_df[race_df['Driver'] == driver].copy()

      # Add the DataFrame to the dictionary with the stint number as the key
      driver_df[driver] = driver_df_temp.reset_index(drop=True)

  driver_stint_split = {}

  for driver in drivers:
      stints = driver_df[driver]['Stint'].unique()
      stint_df = {}
      for stint in stints:
          # Filter the DataFrame for the current stint number
          stint_df_temp = driver_df[driver][driver_df[driver]['Stint'] == stint].copy()

          # Add the DataFrame to the dictionary with the stint number as the key
          stint_df[stint] = stint_df_temp.reset_index(drop=True)
      driver_stint_split[driver] = stint_df

  compound_data = pd.DataFrame(columns=['DriverName','NoOfLaps', 'LapTime', 'Stint', 'Compound'],index=[])

  for driver in drivers:
      stints = driver_df[driver]['Stint'].unique()
      for stint in stints:
          if (compound == driver_stint_split[driver][stint]['Compound']).all():
              driverName = driver
              lapNumber = driver_stint_split[driver][stint]['LapTime'].count()
              lapTime = driver_stint_split[driver][stint]['LapTime'].mean()
              stintNo = stint
              compoundUsed = compound

              compound_data_temp = pd.DataFrame({'DriverName': [driverName],
                                                  'NoOfLaps': [lapNumber],
                                                  'LapTime': [lapTime],
                                                  'Stint': [stintNo],
                                                  'Compound': [compoundUsed]})

              # Append the stint regression data to the overall regression_data DataFrame
              compound_data = pd.concat([compound_data, compound_data_temp], ignore_index=True)

  driverStint = []
  for index , row in compound_data.iterrows():
    stint = int(row['Stint'])
    driverStint.append(row['DriverName']+"@" +str(stint))

  compound_data['Driver+Stint'] = driverStint


  return compound_data
```

```python
def plot_compound_analysis(compound_data):
  import plotly.express as px
  # Sort the DataFrame by LapTime in ascending order
  compound_data_sorted = compound_data.sort_values('LapTime', ascending=True)

  # Create the bar graph using Plotly Express
  fig = px.bar(compound_data_sorted, x='Driver+Stint', y='LapTime', color='Stint', barmode='group',
              labels={'LapTime': 'Lap Time', 'DriverName': 'Driver Name', 'Stint': 'Stint'},
              text='NoOfLaps')

  # Set the title
  fig.update_layout(title='Avg LapTime by All the drivers', barmode='group')

  # Add labels to each bar
  fig.update_traces(texttemplate='%{text}', textposition='outside')

  fig.update_traces(width=0.4, marker=dict(line=dict(width=0.3)))

  # Show the plot
  fig.show()
```

```python
def stint_data(StintNo):
  race_df = race.laps.reset_index()

  race_df['LapTime'] = race_df['LapTime'].dt.total_seconds()

  race2_df = race_df[(race_df['PitInTime'].isnull()) & (race_df['PitOutTime'].isnull())].copy()


  drivers = race_df['Driver'].unique()

  driver_df = {}

  # Iterate over the unique stint numbers
  for driver in drivers:
      # Filter the DataFrame for the current stint number
      driver_df_temp = race_df[race_df['Driver'] == driver].copy()

      # Add the DataFrame to the dictionary with the stint number as the key
      driver_df[driver] = driver_df_temp.reset_index(drop=True)

  driver_stint_split = {}

  for driver in drivers:
      stints = driver_df[driver]['Stint'].unique()
      stint_df = {}
      for stint in stints:
          # Filter the DataFrame for the current stint number
          stint_df_temp = driver_df[driver][driver_df[driver]['Stint'] == stint].copy()

          # Add the DataFrame to the dictionary with the stint number as the key
          stint_df[stint] = stint_df_temp.reset_index(drop=True)
      driver_stint_split[driver] = stint_df


  stint_data = pd.DataFrame(columns=['DriverName','NoOfLaps', 'LapTime', 'Stint', 'Compound'],index=[])


  for driver in drivers:
      stints = driver_df[driver]['Stint'].unique()
      for stint in stints:
          if (StintNo == driver_stint_split[driver][stint]['Stint']).all():
              driverName = driver
              lapNumber = driver_stint_split[driver][stint]['LapTime'].count()
              lapTime = driver_stint_split[driver][stint]['LapTime'].mean()
              stintNo = stint

              compoundUsed = driver_stint_split[driver][stint]['Compound'].iloc[0]

              stint_data_temp = pd.DataFrame({'DriverName': [driverName],
                                                  'NoOfLaps': [lapNumber],
                                                  'LapTime': [lapTime],
                                                  'Stint': [stintNo],
                                                  'Compound': [compoundUsed]})

              # Append the stint regression data to the overall regression_data DataFrame
              stint_data = pd.concat([stint_data, stint_data_temp], ignore_index=True)


  return stint_data
```

```python
import plotly.express as px

def plot_stint_analysis(data):
    # Sort the DataFrame by LapTime in ascending order
    data_sorted = data.sort_values('LapTime', ascending=True)

    color_mapping = {
        'SOFT': 'red',
        'MEDIUM': 'yellow',
        'HARD': 'white'
    }

    # Create the bar graph using Plotly Express
    fig = px.bar(data_sorted, x='DriverName', y='LapTime',color = 'Compound' , color_discrete_map=color_mapping,
                 labels={'LapTime': 'Lap Time', 'DriverName': 'Driver Name'},
                 text='NoOfLaps')

    # Set the title
    fig.update_layout(title='Average Lap Time by Drivers')

    # Add labels to each bar
    fig.update_traces(texttemplate='%{text}', textposition='outside')

    fig.update_layout(xaxis={'categoryorder': 'array', 'categoryarray': data_sorted['DriverName']})

    fig.update_traces(width=0.4, marker=dict(line=dict(width=0.3)))


    # Show the plot
    fig.show()
```

```python
def stint_analysis(stint):
  df = stint_data(stint)
  plot_stint_analysis(df)
```
